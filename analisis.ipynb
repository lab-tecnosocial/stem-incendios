```python
import pandas as pd
```

```python
df = pd.read_csv('Data_clima_Clasificaction.csv')
```

```python
df
```

```python
import re

# Crear un diccionario para almacenar las columnas a combinar
columns_to_combine = {}

# Identificar las columnas que tienen el mismo nombre base
for col in df.columns:
    base_name = re.sub(r'_\d+$', '', col)
    if base_name not in columns_to_combine:
        columns_to_combine[base_name] = []
    columns_to_combine[base_name].append(col)

# Crear un nuevo DataFrame con las columnas combinadas
combined_df = pd.DataFrame()

for base_name, cols in columns_to_combine.items():
    if len(cols) > 1:
        combined_df[base_name] = df[cols].mean(axis=1)
    else:
        combined_df[base_name] = df[cols[0]]

combined_df
```

```python
# save combined_df to a new csv file
combined_df.to_csv('data_11.csv', index=False)
```
